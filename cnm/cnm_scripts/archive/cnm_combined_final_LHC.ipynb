# Charmonia CNM Combined Analysis: LHC p+Pb 5.02 & 8.16 TeV

This notebook performs the full unified CNM analysis (nPDF + Energy Loss + pT Broadening) for Charmonia (J/psi) at LHC energies.

It uses the `CNMCombine` framework to ensure consistency with RHIC analyses.

```python
import sys
from pathlib import Path
import matplotlib.pyplot as plt

# Add parent directory to path
sys.path.append("..")

from cnm_combine.cnm_combine import CNMCombine

# Energies to analyze
ENERGIES = ["5.02", "8.16"]
```

## Run Production for All Energies

The following loop initializes the CNM physics for each energy and generates:
1. $R_{pA}$ vs Rapidity ($y$)
2. $R_{pA}$ vs Transverse Momentum ($p_T$)
3. $R_{pA}$ vs Centrality

```python
for energy in ENERGIES:
    print(f"=== Starting Analysis for p+Pb {energy} TeV ===")
    
    # 1. Initialize System
    cnm = CNMCombine.from_defaults(energy)
    print(f"  Loaded System: {cnm.spec.system}")
    print(f"  Target A: {cnm.spec.A}, Sigma_NN: {cnm.spec.sigma_nn_mb} mb")
    
    # 2. Run Full Production (Calculations + Plots + Saving)
    # This will save CSVs and PDFs to outputs/LHC/CNM_{energy}
    cnm.run_and_save_production()
    
    print(f"=== Finished {energy} TeV ===\n")
```
