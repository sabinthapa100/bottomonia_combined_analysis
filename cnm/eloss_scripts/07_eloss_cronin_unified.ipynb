# Unified Eloss & Cronin Analysis (LHC & RHIC)

This notebook unifies the analysis for pPb LHC (5.02 TeV) and dAu RHIC (200 GeV) systems.
It uses the extracted configurations and the new `GlauberWrapper` to ensure consistency and replicability.

**Key Features:**
*   Uses `eloss_code/system_configs.py` for all parameters.
*   Uses `eloss_code/glauber_wrapper.py` for Glauber interface.
*   Replicates 'Gold Standard' plots from `05_b` and `06_b`.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Add code directory to path
sys.path.append(os.path.abspath("../"))

from eloss_code.system_configs import LHCConfig, RHICConfig
from eloss_code.glauber_wrapper import GlauberWrapper
import eloss_code.quenching_fast as QF

# Ensure consistent plotting style
plt.style.use('seaborn-v0_8-whitegrid') # Or similar
```

## 1. System Initialization
Initialize Glauber model wrappers for both systems using the extracted configurations.

```python
# LHC Initialization
print("Initializing LHC Glauber...")
lhc_spec = LHCConfig.spec_5TeV
gl_lhc = GlauberWrapper(lhc_spec.system, lhc_spec.roots_GeV, lhc_spec.A, lhc_spec.sigma_nn_mb)

# RHIC Initialization
print("Initializing RHIC Glauber...")
rhic_spec = RHICConfig.spec
gl_rhic = GlauberWrapper(rhic_spec.system, rhic_spec.roots_GeV, rhic_spec.A, rhic_spec.sigma_nn_mb)
```

## 2. Configuration Verification
Print out key parameters to verify they match instructions.

```python
print("--- LHC Configuration ---")
print(f"Sqrt(s): {LHCConfig.roots5} GeV")
print(f"Centrality Bins (Plotting): {LHCConfig.cent_bins_plotting}")
print(f"Rapidity Windows: {LHCConfig.rapidity_windows}")
print(f"pT Edges: {LHCConfig.pT_edges}")

print("\n--- RHIC Configuration ---")
print(f"Sqrt(s): {RHICConfig.roots} GeV")
print(f"Centrality Bins: {RHICConfig.cent_bins}")
print(f"Rapidity Windows: {RHICConfig.rapidity_windows}")
print(f"pT Edges: {RHICConfig.pT_edges}")
print(f"Custom pp Params: {RHICConfig.pp_params}")
```

## 3. Analysis Logic Placeholder
This section will contain the logic to generate the plots using `quenching_fast` routines,
adapted to use `GlauberWrapper` and the config objects.

```python
# Example: Calculate Leff for LHC centrality bins
print("Calculating Leff for LHC bins...")
leff_lhc = gl_lhc.leff_bins_pA(LHCConfig.cent_bins_plotting)
for label, val in leff_lhc.items():
    print(f"  {label}: {val:.4f} fm")

# Example: Calculate Leff for RHIC centrality bins
print("\nCalculating Leff for RHIC bins...")
leff_rhic = gl_rhic.leff_bins_pA(RHICConfig.cent_bins) # Using the wrapper's unified method
for label, val in leff_rhic.items():
    print(f"  {label}: {val:.4f} fm")
```
